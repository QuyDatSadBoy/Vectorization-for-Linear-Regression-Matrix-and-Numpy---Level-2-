```python
import numpy as np

data = np.genfromtxt('advertising.csv', 
                     delimiter=',', 
                     skip_header=1)
print(data.shape)
```

```python
N = 150
n_epochs = 50
lr = 0.00001

# khởi tạo giá trị tham số
thetas = np.random.randn(4,)
losses = []

for epoch in range(n_epochs):
    for i in range(N):
        # get 1 sample          
        sample_i = data[i]
        
        # convert to vector
        x_1 = sample_i[0]
        x_2 = sample_i[1]
        x_3 = sample_i[2]
        y_i = sample_i[3]
        vector_x = np.array([1, x_1, x_2, x_3])
        
        # compute output 
        y_hat = vector_x.T.dot(thetas)

        # compute loss li
        loss = (y_hat - y_i)*(y_hat - y_i)

        # compute gradient 
        gradients = 2*(y_hat - y_i)*vector_x

        # update 
        thetas = thetas - lr*gradients

        # logging          
        losses.append(loss)
```

```python
import matplotlib.pyplot as plt

# in loss cho 500 sample đầu
x_axis = list(range(500))
plt.plot(x_axis,losses[:500], color="r")
plt.show()
```
